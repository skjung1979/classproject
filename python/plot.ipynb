```python
# wordcloud
import wordcloud
text = 'python python hello data data data data data data'
wc = wordcloud.WordCloud()
wc.generate_from_text(text)
wc.to_image()
```

```python
text = {'python':10, 'hello':1, 'data':5}
wc = wordcloud.WordCloud()
wc.generate_from_frequencies(text)
wc.to_image()
```

```python
text = '파이썬 파이썬 파이썬 시각화 데이터분석 오류 데이터'
font_path='/windows/fonts/batang.ttc'

# 시각화 이미지 설정: 한글 폰트, 배경색
wc = wordcloud.WordCloud(font_path=font_path, background_color='pink', stopwords=['오류','데이터']) #생략 키워드 설정:stopwords
wc.generate_from_text(text)
wc.to_image()
```

```python
import numpy as np
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,2,3,4]
y = [2,4,6,8]

x2 = np.array([1,2,3,4])
y2 = np.array([4,1,3,7])

axes.plot(x,y, color='red', linestyle='solid', marker='o')
axes.plot(x2,y2, color='green', linestyle='dashed', marker='^')
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,3,5,7]
y = [2,4,6,8]

x2 = [2,4,6,8]
y2 = [4,4,3,6]

axes.bar(x,y, color='blue')
axes.bar(x2,y2, color='green')
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)
axes2 = axes.twinx()

x = [1,2,3,4]
y = [2,4,6,8]

x2 = [1,2,3,4]
y2 = [4,6,1,9]

axes.bar(x,y, color='blue', label='bar')
axes2.plot(x2,y2, color='r', label='plot')

axes.legend()
axes2.legend(loc=1)

plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)


x = [1,2,3,4]
y = [2,4,6,8]

x2 = [1,2,3,4]
y2 = [4,6,1,9]

axes.scatter(x,y, color='blue')
axes.scatter(x2,y2, color='green')

plt.show()
```
